linear equation(선형 방정식)은 
$$
a_{1}x_{1} + a_{2}x_{2} + \cdots + a_{n}x_{n} = b
$$         
형태로 되어있으며, 이런 linear equation의 solution은 
1. solution이 존재하지 않거나
2. solution이 단 한가지만 존재하거나
3. solution이 무한하게 존재하거나

3가지 경우로 나올 수 있다. 이때 solution이 존재하는 경우를 consist하다고 얘기한다.

선형방정식은 matrix(행렬)형식으로 만들 수 있는데, 계수만 있는 행렬을 coefficient matrix라고 부르고, 계수와 상수가 있는 행렬을 agumented matrix라고 부른다.

##(정의1)
직사각형 행렬들은 echelon form을 띄는데 만약 3가지의 속성을 가지고 있다면 echelon matrix라고 부른다. ex) (1)
1. 0이 아닌 모든 행이 0인 모든 행들보다 위에 있을 때
2. 행의 각 선행 요소는 그 위에 있는 행의 선행 항목 오른쪽에 있는 열에 있을 때
3. 선행 요소 아래 열의 모든 요소들이 0일 때

만약 echelon form을 뛰는 matrix가 밑의 두가지의 상태를 가지면 그것은 reduce echelon form이라고 부른다. ex) (2)
4. 0이 아닌 행의 선행 요소가 1일 때
5. 각 선행 1이 그 열 안에서 오직 0이 아닌 요소일 때

$$
\begin{bmatrix}
2 & -3 & 2 & 1\\ 
0 & 1 & -4 & 8\\ 
0 & 0 & 0 & 5/2
\end{bmatrix}
(1)
\quad
\begin{bmatrix}
1 & 0 & 0 & 29\\ 
0 & 1 & 0 & 16\\ 
0 & 0 & 1 & 3
\end{bmatrix}
(2)
$$

##(정리1)<-(정의1)
###Reduced echelon form의 고유성
각 행렬은 1과 row equivalent(행상등)하고 오직 하나의 reduce echelon matrix를 갖는다.

##(정의2)
matrix A의 pivot position은 A의 reduced echelon form안에 선행 1에 해당하는 위치에 있다. pivot column은 pivot position을 포함하는 A의 열을 pivot column이라고 부른다.
ex)즉 밑에 있는 행렬에서 pivot position은 첫 번째 행 첫 번째 열의 값인 2이고 pivot column은 pivot position이 있는 열이다.
$$
\begin{bmatrix}
2 & -3 & 2 & 1\\ 
0 & 1 & -4 & 8\\ 
0 & 0 & 0 & 5/2
\end{bmatrix}
$$

##(정리2)<-(정의2)
###존재 그리고 고유성의 정리
선형 방정식이 consistent이다와  augmented matrix의 맨 오른쪽의 열이 pivot column이 아니다는 필요충분조건이다. (선형 방정식이 consistent하면 해가 존재하는데 맨 오른쪽의 열이 pivot column이 되면 pivot position의 앞에 값들이 모두 0이되면서 coefficient값들은 0이고 계수만 b가 되므로 이 경우 해가 존재하지 않는다.)
$$
\begin{bmatrix}
0 & \cdots & 0 & b
\end{bmatrix}
\quad
with \;b \;nonzero
$$

##(정의3)
$v_{1}, ..., v_{p}$가 $\mathbb{R}^{n}$의 안에있고, $v_{1}, ..., v_{p}$의 모든 선형 결합을 가지는 집합들을 Span한 $\mathbb{R}^{n}$의 하위집합이라고 부른다.

##(정의4)
만약 A가 $m\times n$일때, 열은 $a_{1}, ..., a_{n}$, 그리고 만약 x가 $\mathbb{R}^{n}$안에 있다고 할때, Ax로 표시되는 A와 x의 곱은 x의 해당 항목을 가중치로 사용하는 A 열의 선형 조합이다.
$$ A\mathbf{x} = \begin{bmatrix}a_{1} & a_{2} & \cdots & a_{n}\end{bmatrix} \begin{bmatrix} x_{1}\\ \vdots \\ x_{n} \end{bmatrix} = x_{1}a_{1} + x_{2}a_{2} + \cdots + x_{n}a_{n}$$

##(정리3)<-(정의4)
A가 $m\times n$이고, 열은 $a_{1}, ...,a_{n}$ 그리고 만약 b가 $\mathbb{R}^{n}$안에 있다고 할 때, matrix equation $A\mathbf{x} = b$은 vector equation과 같은 solution set을 가진다. 같은 solution set을 가지면서 선형방정식의 agumented matrix의 모습은 $\begin{bmatrix}a_{1} & a_{2} & \cdots & a_{n} & b\end{bmatrix}$이다.

##(정리4)<-(정의3)(정의2)(정리2)(정의4)
A가 $m\times n$일 때, 다음의 문장들은 논리적으로 동일하다. 즉, 특정한 A에서 모두 참이거나, 모두 거짓이다.
1. 각 b가 $\mathbb{R}^{m}$에 있을 때, 방정식 $A\mathbf{x} = b$는 solution을 가진다.
2. $\mathbb{R}^{m}$안에 각 b는 A의 열과 선형 결합이다. 
3. A의 열은 $\mathbb{R}^{m}$의 Span이다. 
4. A는 모든 행에서 pivot position을 가진다.

##(정리5)
만약 A가 $m\times n$이고, $\mathbf{u}$와 $\mathbf{v}$가 vector $\mathbb{R}^{n}$에 속하고, $c$는 스칼라이면 
1. $A(\mathbf{u} + \mathbf{v}) = A\mathbf{u} + A\mathbf{v}$
2. $A(c\mathbf{u}) = c(A\mathbf{u})$

##(정리6)
방정식 $A\mathbf{x} = \mathbf{b}$가 consistent하고 주어진 $\mathbf{b}$에 대해서 solution으로 $\mathbf{p}$가 주어진다고 가정할 때, $A\mathbf{x} = \mathbf{b}$의 solution 집합은 $\mathbf{w} = \mathbf{p}+\mathbf{v}_{h}$ 형식의 모든 vector의 집합이다. 여기서 $\mathbf{v}_{h}$는 homogeneous equation인 Ax=0의 해이다.

##(정의5)
인덱스 된 vector 집합 $\begin{Bmatrix}\mathbf{v}_{1}, ... ,\mathbf{v}_{p}\end{Bmatrix}$, $\mathbb{R}^{n}$에서 다음과 같은 벡터 방정식이 오직 trivial solution만을 가질때 linearly independent하다고 한다.
$$x_{1}\mathbf{v}_{1} + x_{2}\mathbf{v}_{2} + \cdots + x_{p}\mathbf{v}_{p} = 0$$
모두 0이 아닌 가중치 $c_{1},...,c_{p}$가 존재할 때, 집합$\begin{Bmatrix}\mathbf{v}_{1}, ..., \mathbf{v}_{p}\end{Bmatrix}$ linearly dependent라고 부른다. 밑의 식과 같이,
$$c_{1}\mathbf{v}_{1}+c_{2}\mathbf{v}_{2}+\cdots+c_{p}\mathbf{v}_{p}=0$$

##(정리7)<-(정의5)
###Linearly dependent 집합의 특징
둘 이상의 vector로 되어진 집합 $S = \begin{Bmatrix}\mathbf{v}_{1}, ..., \mathbf{v}_{p}\end{Bmatrix}$이 linearly dependent이다 와 적어도 $S$안에 있는 하나의 vector는 다른 vector의 선형 결합이다는 필요충분조건이다.

##(정리8)
집합에 각 vector의 요소보다 더 많은 vector가 포함되어 있으면 집합은 linearly dependent이다. 즉 $\mathbb{R}^{n}$에 속해 있는 어떤 집합 $\begin{Bmatrix}\mathbf{v}_{1}, ...,\mathbf{v}_{p}\end{Bmatrix}$가 $p>n$일 때, linearly dependent하다.

##(정리9)<-(정의5)
만약 $\mathbb{R}^{n}$안의 집합 $S = \begin{Bmatrix}\mathbf{v}_{1}, ..., \mathbf{v}_{p}\end{Bmatrix}$가 0 vector를 포함하면, 그 집합은 linearly dependent하다.

##(정의6)
transformation(or mapping)$T$는 다음과 같은 경우에 linear하다:
1. $T(\mathbf{u}+\mathbf{v})=T(\mathbf{u})+T(\mathbf{v})$인 모든 $\mathbf{u},\mathbf{v}$는 $T$의 영역안에 있을 때
2. $T(c\mathbf{u}) = cT(\mathbf{u})$인 모든 스칼라 $c$그리고 모든 $\mathbf{u}$가 $T$의 영역안에 있을 때

##(정리10)<-(정의4)
$T:\mathbb{R}^{n}\rightarrow \mathbb{R}^{m}$는 linear transformation일 때, $A$와 같은 unique한 matrix가  존재한다. 
$$T(\mathbf{x}) = A\mathbf{x} \; for\; all\; \mathbf{x}\; in\; \mathbb{R}^{n}$$
$A$는 $j$번째 열이 vector $T(\mathbf{e}_{j})$인  $m\times n$ matrix이고, 여기서 $\mathbf{e}_{j}$는 identity matrix인 $\mathbb{R}^{n}$의 $j$번째 열이다.
$$A = \begin{bmatrix}T(\mathbf{e}_{1}) & ... & T(\mathbf{e}_{n})\end{bmatrix}$$

##(정의7)
만약 $\mathbb{R}^{m}$의 각 $\mathbf{b}$가 $\mathbb{R}^{n}$에 있는 적어도 하나 이상의 $\mathbf{x}$ 가 있을때, mapping $T:\mathbb{R}^{n}\rightarrow \mathbb{R}^{m}$는 onto$\mathbb{R}_{m}$ 하다고 한다.

##(정의8)
만약 $\mathbb{R}^{m}$의 각 $\mathbf{b}$가 $\mathbb{R}^{n}$의 최대 하나의 $\mathbf{x}$를 가질 때, mapping $T:\mathbb{R}^{n}\rightarrow \mathbb{R}^{m}$는 one-to-one하다고 한다.

##(정리11)<-(정의6)(정의8)
$T:\mathbb{R}^{n}\rightarrow \mathbb{R}^{m}$가 linear transformation일 때, $T$는 one-to-one이다 와 방정식 $T(\mathbf{x}) = 0$은 오직 trivial solution($\mathbf{x}$가 zero vector만을 가지고 있는 경우)을 갖는다는 필요충분조건이다.

##(정리12)<-(정리4)(정리11)
$T:\mathbb{R}^{n}\rightarrow \mathbb{R}^{m}$가 linear transformation이고, $A$가 $T$에 standard matrix이면:
1. $T$ 는 $\mathbb{R}^{n}$는 onto$\mathbb{R}^{m}$이다 와, A의 열은 Span $\mathbb{R}^{m}$이다는 필요충분조건이다.
2. $T$는 one-to-one이다와 $A$의 열은 linearly independent하다는 필요충분 조건이다.
